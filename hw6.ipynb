###  Find the unlabeled data point that is the farthest to the decision boundary
### of the SVM. Let the SVM label it (ignore its true label), and add it to
### the labeled data, and retrain the SVM. Continue this process until all
### unlabeled data are used.

```python
unlabel
```

```python
import numpy as np
```

```python
# ref: https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC.decision_function
# ref: https://stackoverflow.com/questions/32074239/sklearn-getting-distance-of-each-point-from-decision-boundary
decision = clf.decision_function(unlabel)
w_norm = np.linalg.norm(clf[-1].coef_)
dist = decision / w_norm
```

```python
# get the farthest and predict with our model
farthest_index = np.argmax(np.absolute(dist))
predicted_label = clf.predict(unlabel.iloc[[farthest_index]])
selected_row = unlabel.iloc[[farthest_index]]
```

```python
selected_row
```

```python
# add the selected data to train sample
X_train_sample = X_train_sample.append(selected_row, ignore_index=True)
```

```python
X_train_sample
```

```python
# add predicted label to true label
y_train_sample = y_train_sample.append(pd.DataFrame(pd.DataFrame(predicted_label, columns=['label'])), ignore_index=True)
```

```python
y_train_sample
```

```python
# normalize train sample
norm_X_train_sample = preprocessing.normalize(X_train_sample, norm='l2')
norm_X_train_sample
```

```python
# retrain svm
clf.fit(norm_X_train_sample, y_train_sample)
```

```python
unlabel = unlabel.drop([farthest_index]).reset_index(drop=True)
```

```python
remain_steps = len(unlabel)
remain_steps
```

```python
# continue above processes until every unlabel data is gone
for i in range(remain_steps):
    # find the farthest data point
    decision = clf.decision_function(unlabel)
    w_norm = np.linalg.norm(clf[-1].coef_)
    dist = decision / w_norm
    farthest_index = np.argmax(np.absolute(dist))
    
    # predict on the selected data
    predicted_label = clf.predict(unlabel.iloc[[farthest_index]])
    selected_row = unlabel.iloc[[farthest_index]]
    
    # add the selected data to train sample
    X_train_sample = X_train_sample.append(selected_row, ignore_index=True)
    # add predicted label to true label
    y_train_sample = y_train_sample.append(pd.DataFrame(pd.DataFrame(predicted_label, columns=['label'])), ignore_index=True)
    
    # normalize train sample
    norm_X_train_sample = preprocessing.normalize(X_train_sample, norm='l2')
    
    # retrain
    clf.fit(norm_X_train_sample, y_train_sample)
    
    # drop the data from unlabel
    unlabel = unlabel.drop([farthest_index]).reset_index(drop=True)
```

```python
    target_names = ['class B', 'class M']
    ## report on train
    y_pred = clf.predict(norm_X_train_sample)
    y_true = y_train_sample
    print(classification_report(y_true, y_pred, target_names=target_names))
```

```python
for m in range(30):  
    
    sample_M = class_M.sample(n = 106 ) 
    sample_B = class_B.sample(n = 179 ) 
    # randomly select unlabeled data
    unlabel_M = class_M.drop(sample_M.index)
    unlabel_B = class_B.drop(sample_B.index)
    
    # concate pos and neg classses for labeled data
    train_sample = sample_M.append(sample_B, ignore_index=True)
    X_train_sample = train_sample.iloc[:, 1:22]
    y_train_sample = train_sample.iloc[:, 0:1]

    # concate dataframes for unlabeled data; drop label
    unlabel = unlabel_M.append(unlabel_B, ignore_index=True)
    unlabel = unlabel.iloc[:, 1:22]
    
    # normalize trian X
    norm_X_train_sample = preprocessing.normalize(X_train_sample, norm='l2')
    
    # Train an L1-penalized SVM t with selected penalty parameter
    clf = make_pipeline(StandardScaler(), LinearSVC(C=1, penalty='l1',dual=False))
    clf.fit(norm_X_train_sample, y_train_sample)
    
    remain_steps = len(unlabel)
    
    for i in range(remain_steps):
        # find the farthest data point
        decision = clf.decision_function(unlabel)
        w_norm = np.linalg.norm(clf[-1].coef_)
        dist = decision / w_norm
        farthest_index = np.argmax(np.absolute(dist))
    
        # predict on the selected data
        predicted_label = clf.predict(unlabel.iloc[[farthest_index]])
        selected_row = unlabel.iloc[[farthest_index]]
    
        # add the selected data to train sample
        X_train_sample = X_train_sample.append(selected_row, ignore_index=True)
        # add predicted label to true label
        y_train_sample = y_train_sample.append(pd.DataFrame(pd.DataFrame(predicted_label, columns=['label'])), ignore_index=True)
    
        # normalize train sample
        norm_X_train_sample = preprocessing.normalize(X_train_sample, norm='l2')
    
        # retrain
        clf.fit(norm_X_train_sample, y_train_sample)
    
        # drop the data from unlabel
        unlabel = unlabel.drop([farthest_index]).reset_index(drop=True)
    
    target_names = ['class B', 'class M']
    ## report on train
    y_pred = clf.predict(norm_X_train_sample)
    y_true = y_train_sample
    print(classification_report(y_true, y_pred, target_names=target_names))
```

```python
def majorityElement(label_list):
    count_M=0
    count_B=0
    for i in range(len(label_list)):
        if(label_list[i]=='M'):
            count_M += 1
        elif(label_list[i]=='B'):
            count_B += 1
            
    if(count_M>=count_B):
        return 'M'
    else:
        return 'B'
```

```python
cluster_1_label=majorityElement(cluster_1_true)
cluster_2_label=majorityElement(cluster_2_true)
```

```python
y_pred_label = kmeans.labels_
y_pred = []
for i in range(len(y_pred_label)):
    if(y_pred_label[i]==0):
        y_pred.append(cluster_1_label)
    else:
        y_pred.append(cluster_2_label)
```

### (b) Repeat each of the following two procedures 50 times. You will have 50 errors for
#### 90 SVMs per each procedure.

 ### You will have 90 SVMs that were trained using 10, 20, 30, ... , 900 data points and their 90 test erro

```python
from sklearn.metrics import mean_squared_error
```

```python
test_error_lst = []
train_size_lst = []
for i in range(10, 910,10):
    train_X = train_2_X.iloc[0:i] ## 10, 20, 30, ... , 900 for each iteration
    train_y = train_2_y.iloc[0:i]
    
   
    
    
    parameters = {'C':[0.001,0.01,0.1,1,10,100]}
    svc = LinearSVC(penalty='l1',dual=False)
    clf = GridSearchCV(svc, parameters, cv=5)
    clf.fit(train_X, train_y)
    
    c= float(clf.best_params_['C'])
    
    clf = make_pipeline(StandardScaler(), LinearSVC(C=c, penalty='l1',dual=False))
    clf.fit(train_X, train_y)
    
    y_pred = clf.predict(test_2_X)
    y_true = test_2_y
    
    print ("Test error for ",i," :",mean_squared_error(y_true, y_pred))
    test_error_lst.append(mean_squared_error(y_true, y_pred))
    train_size_lst.append(i)
```

```python
import matplotlib.pyplot as plt
plt.plot(train_size_lst, test_error_lst)
plt.xlabel('train size')
plt.ylabel('test error')
plt.title('Passive learning')
plt.show()
```

```python
train_2_X_re = train_2_X.reset_index(drop=True) ## 10, 20, 30, ... , 900 for each iteration
train_2_y_re = train_2_y.reset_index(drop=True)

train_X = train_2_X_re.iloc[0:10] 
train_y = train_2_y_re.iloc[0:10]

lst_active = []

for i in range(10, 910,10):
    
    parameters = {'C':[0.001,0.01,0.1,1,10,100]}
    svc = LinearSVC(penalty='l1',dual=False)
    clf = GridSearchCV(svc, parameters, cv=5)
    clf.fit(train_X, train_y)
    
    c= float(clf.best_params_['C'])
    
    clf = make_pipeline(StandardScaler(), LinearSVC(C=c, penalty='l1',dual=False))
    clf.fit(train_X, train_y)
    
    decision = clf.decision_function(train_2_X)
    w_norm = np.linalg.norm(clf[-1].coef_)
    dist = decision / w_norm
    
    closest_10_index = []
    selected_data = []
    for j in range(10):
        if(i==10):
            break
        else:
            closest_10_index.append(np.argmax(np.absolute(dist)))
            dist[closest_10_index[j]] = 0
            selected_data = train_2_X_re.iloc[[closest_10_index[j]]]
            selected_label = train_2_y_re.iloc[[closest_10_index[j]]]
            train_X = train_X.append(selected_data, ignore_index=True)
            train_y = train_y.append(selected_label, ignore_index=True)
                
    
    clf = make_pipeline(StandardScaler(), LinearSVC(C=c, penalty='l1',dual=False))
    clf.fit(train_X, train_y)
    
    y_pred = clf.predict(test_2_X)
    y_true = test_2_y
    
    print ("Test error for ",i," :",mean_squared_error(y_true, y_pred))
    lst_active.append(mean_squared_error(y_true, y_pred))
```

```python
import matplotlib.pyplot as plt
plt.plot(train_size_lst, lst_active)
plt.xlabel('train size')
plt.ylabel('test error')
plt.title('Active learning')
plt.show()
```
